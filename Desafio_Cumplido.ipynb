### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingresos totales sumando la columna 'Precio'
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostramos el resultado
print(f"Ingreso total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso_tienda4:,.2f}")


```

# 2. Ventas por categoría

```python
# Cantidad de productos vendidos por categoría en Tienda 1
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría - Tienda 1:")
print(ventas_categoria_tienda1)
print("\n")

# Cantidad de productos vendidos por categoría en Tienda 2
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría - Tienda 2:")
print(ventas_categoria_tienda2)
print("\n")

# Cantidad de productos vendidos por categoría en Tienda 3
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría - Tienda 3:")
print(ventas_categoria_tienda3)
print("\n")

# Cantidad de productos vendidos por categoría en Tienda 4
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría - Tienda 4:")
print(ventas_categoria_tienda4)
print("\n")
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio de clientes en Tienda 1
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")

# Calificación promedio de clientes en Tienda 2
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")

# Calificación promedio de clientes en Tienda 3
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")

# Calificación promedio de clientes en Tienda 4
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Productos más y menos vendidos - Tienda 1
productos_tienda1 = tienda['Producto'].value_counts()
print("Producto más vendido Tienda 1:")
print(productos_tienda1.head(1))  # Top 1
print("\nProducto menos vendido Tienda 1:")
print(productos_tienda1.tail(1))  # Bottom 1
print("\n")

# Productos más y menos vendidos - Tienda 2
productos_tienda2 = tienda2['Producto'].value_counts()
print("Producto más vendido Tienda 2:")
print(productos_tienda2.head(1))
print("\nProducto menos vendido Tienda 2:")
print(productos_tienda2.tail(1))
print("\n")

# Productos más y menos vendidos - Tienda 3
productos_tienda3 = tienda3['Producto'].value_counts()
print("Producto más vendido Tienda 3:")
print(productos_tienda3.head(1))
print("\nProducto menos vendido Tienda 3:")
print(productos_tienda3.tail(1))
print("\n")

# Productos más y menos vendidos - Tienda 4
productos_tienda4 = tienda4['Producto'].value_counts()
print("Producto más vendido Tienda 4:")
print(productos_tienda4.head(1))
print("\nProducto menos vendido Tienda 4:")
print(productos_tienda4.tail(1))
print("\n")
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio en Tienda 1
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
print(f"Costo de envío promedio Tienda 1: ${costo_envio_promedio_tienda1:.2f}")

# Costo de envío promedio en Tienda 2
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
print(f"Costo de envío promedio Tienda 2: ${costo_envio_promedio_tienda2:.2f}")

# Costo de envío promedio en Tienda 3
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
print(f"Costo de envío promedio Tienda 3: ${costo_envio_promedio_tienda3:.2f}")

# Costo de envío promedio en Tienda 4
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
print(f"Costo de envío promedio Tienda 4: ${costo_envio_promedio_tienda4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Datos de ventas por categoría en cada tienda
categorias = ['Muebles', 'Electrónicos', 'Juguetes', 'Electrodomésticos', 'Deportes y diversión', 'Instrumentos musicales', 'Libros', 'Artículos para el hogar']

# Ventas por categoría para cada tienda
ventas_tienda1 = [465, 448, 324, 312, 284, 182, 173, 171]
ventas_tienda2 = [442, 422, 313, 305, 275, 224, 197, 181]
ventas_tienda3 = [499, 451, 315, 278, 277, 185, 177, 177]
ventas_tienda4 = [480, 451, 338, 277, 254, 201, 187, 170]

# Crear gráficos de pastel para cada tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ventas_por_tienda = [ventas_tienda1, ventas_tienda2, ventas_tienda3, ventas_tienda4]

for i, ventas in enumerate(ventas_por_tienda):
    plt.figure(figsize=(8, 8))
    plt.pie(ventas, labels=categorias, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribución de Ventas por Categoría - {tiendas[i]}')
    plt.axis('equal')  # Asegura que el gráfico sea un círculo
    plt.show()
```
